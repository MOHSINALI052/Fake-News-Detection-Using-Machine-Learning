```python
import re
import string

def wordopt(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove text inside square brackets
    text = re.sub(r'\[.*?\]', '', text)
    
    # Remove non-word characters (but your current "\\w" removes letters, so fixing this)
    text = re.sub(r'\W', ' ', text)
    
    # Remove URLs (http, https, or www links)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove HTML tags
    text = re.sub(r'<.*?>+', '', text)
    
    # Remove punctuation
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    
    # Remove newlines
    text = re.sub(r'\n', '', text)
    
    # Remove words containing digits
    text = re.sub(r'\w*\d\w*', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text



    
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
import pandas as pd

def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news, wordopt, vectorization, LR, DT, GB, RF):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    
    # Apply text processing function (e.g., word optimization)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    
    # Vectorize the processed text
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    
    # Predict using different models
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)
    
    # Combine predictions into a dictionary
    predictions = {
        "Logistic Regression": output_label(pred_LR[0]),
        "Decision Tree": output_label(pred_DT[0]),
        "Gradient Boosting": output_label(pred_GB[0]),
        "Random Forest": output_label(pred_RF[0])
    }
    
    return predictions
```

```python
# Example of the required setup
# Assuming wordopt is a function you defined, and you've trained the models (LR, DT, GB, RF)
news = str(input("Enter news text: "))

# Call the manual_testing function with necessary arguments
predictions = manual_testing(news, wordopt, vectorization, LR, DT, GB, RF)

# Output the predictions
print(predictions)
```
